# Feature Engineering

#### Changing large to '1' and small to '0' into a column

```python

forest['size_category'] = np.where(forest['size_category'].str.contains("large"), 1, 0)
forest.head()
```

#### Dropping month and day column

```python

forest.drop(['month','day'], axis=1, inplace = True)
forest.head()
```

```python
forest.info()
```

```python
X = forest.iloc[:,:28]
Y = forest.iloc[:,28]
```

### Splitting the data into train and test

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3)
```

#### applying Randomized search CV

```python
clf = SVC()
param_grid = [{'kernel':['linear','rbf'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.001,0.0001] }]
GSV = RandomizedSearchCV(clf,param_grid,cv=10,verbose=300)
GSV.fit(x_train,y_train)
```

```python
GSV.best_params_ , GSV.best_score_ 
```

### applying Grid Search CV

```python
clf = SVC()
param_grid = [{'kernel':['linear','rbf'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.001,0.0001] }]
GSV = GridSearchCV(clf,param_grid,cv=10,verbose=300)
GSV.fit(x_train,y_train)
```

```python
GSV.best_params_ , GSV.best_score_
```

### Applying SVM classification

###### In both randomized search CV and grid search CV we are getting the same parameter ('C': 10, 'gamma': 0.5, 'kernel': 'linear') So

```python

clf = SVC(kernel = 'linear', C = 10, gamma = 0.5)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
