# Q4) Two Dice are rolled, find the probability that sum is 

1. Equal to 1 

```python
dice1 =[1,2,3,4,5,6]
dice2 =[1,2,3,4,5,6]

counts=0

for i in range(0,6):
    for j in range(0,6):
        if dice1[i]+dice2[j]==1:
            counts+=1
try:
    print("the probability that sum is equal to 1 is",counts/(len(dice1)*len(dice2)),'%')
except:
    print('the probability is 0%')
```

2. Less than or equal to 4 

```python
counts=0
for i in range(0,6):
    for j in range(0,6):
        if dice1[i]+dice2[j]<=4:
            counts+=1
try:
    print("the probability that sum is less than equal to 4 is",(counts/(len(dice1)*len(dice2)))*100,'%')
except:
    print('the probability is 0 %')
```

3. Sum is divisible by 2 and  3 

```python
counts=0
for i in range(0,6):
    for j in range(0,6):
        if (dice1[i]+dice2[j])%2==0 and (dice1[i]+dice2[j])%3==0:
            counts+=1
try:
    print("the probability that sum is divisible by 2 and 3 is",(counts/(len(dice1)*len(dice2)))*100,'%')
except:
    print('the probability is 0 %')
```

# Q11) Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population  of 3,000,000  men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval? 

```python
mean=200
sd=30
num=2000
from scipy import stats
ci94=stats.norm.interval(0.94,mean,sd)
print('94% confident interval of average male is upto ',np.round(ci94,2))
ci98=stats.t.interval(0.98,num-1,mean,sd)
print('98% confident interval of average male is upto ',np.round(ci98,2))
ci96=stats.t.interval(0.96,num-1,mean,sd)
print('96% confident interval of average male is upto ',np.round(ci96,2))
```

# Q 24)   A Government  company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days # 

```python
# h0 -> n>260
# h1 -> n<260

#Mean=270
#s_mean=260
#std=90
#n=18

t=(260-270)/(90/np.sqrt(18))

pvalue=stats.t.cdf(np.round(t,3),df=17)

print('The probabilty of gitting an average life of no more than 260 days are',np.round((pvalue)*100,3))

```
